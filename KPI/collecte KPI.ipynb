```python
df_par_projet = df.groupby('Nom projet')
```

```python
for Projet, groupe in df_par_projet:
    tickets_estimes = groupe[groupe['Estimation story point'].notnull()].shape[0]
    tickets_non_estimes = groupe[groupe['Estimation story point'].isnull()].shape[0]
    tickets_par_etat = groupe.groupby('État').size().to_dict()
    tickets_par_type = groupe.groupby('Type de ticket').size().to_dict()
    Complexité_par_type = groupe.groupby('Type de ticket').apply(lambda x: x[['Clé ticket', 'Estimation story point']].to_dict('records')).to_dict()
    Complexité_par_etat = groupe.groupby('État').apply(lambda x: x[['Clé ticket', 'Estimation story point']].to_dict('records')).to_dict()
    nombre_bug = groupe[groupe['Type de ticket'] == 'Bug'].shape[0]
    Charge_traitement_bug = groupe[groupe['Type de ticket'] == 'Bug']['Temps consacré'].sum()
    Charge_totale_estimation = groupe['Estimation story point'].sum()
    Pourcentage_charge_traitement = (Charge_traitement_bug / Charge_totale_estimation) * 100
    Nbre_US_clôturés = groupe[(groupe['Type de ticket'] == 'Bug') & (groupe['État'] == 'Terminé')].shape[0]
    Nbre_US_total= groupe[groupe['Type de ticket'] == 'Bug'].shape[0]
    if Nbre_US_total > 0:
        Avancement1 = (Nbre_US_clôturés / Nbre_US_total) * 100
    else:
        Avancement1 = 0 
    Temps_estimé = groupe['Estimation originale'].sum()
    Temps_consommé = groupe['Temps consacré'].sum()
    Avancement2 = (Temps_consommé/Temps_estimé)*100 if Temps_estimé > 0 else 0
    Charge_ticket_estimé = groupe[groupe['Estimation story point'].notnull()]
    # Sélectionner les colonnes pertinentes
    Charge_ticket_estimé = Charge_ticket_estimé[['Clé ticket', 'Estimation story point', 'Temps consacré']]
    # Convertir en liste de dictionnaires
    Charge_ticket_estimé = Charge_ticket_estimé.to_dict(orient='records')
    #charge ticket estimé par statut
    # Calcul de la charge de ticket estimée par état
    charge_ticket_estime_par_etat = groupe[groupe['Estimation story point'].notnull()].groupby('État').apply(lambda x: x[['Clé ticket', 'Estimation story point', 'Temps consacré']].to_dict(orient='records')).to_dict()
    #charge ticket estimé par type
    charge_ticket_estime_par_type = groupe[groupe['Estimation story point'].notnull()].groupby('Type de ticket').apply(lambda x: x[['Clé ticket', 'Estimation story point', 'Temps consacré']].to_dict(orient='records')).to_dict()
    projet_doc = {
        'tickets_estimes': int(tickets_estimes),
        'tickets_non_estimes': int(tickets_non_estimes),
        'Charge ticket estimé': Charge_ticket_estimé,
        'Charge_ticket_estime_par_etat': charge_ticket_estime_par_etat,
        'Charge_ticket_estime_par_type': charge_ticket_estime_par_type,
        'tickets_par_etat': tickets_par_etat,
        'tickets_par_type': tickets_par_type,
        'Complexité_par_type': Complexité_par_type,
        'Complexité_par_etat': Complexité_par_etat,
        'nombre_bug': nombre_bug,
        'Charge_traitement_bug': Charge_traitement_bug,
        'Pourcentage_charge_traitement': Pourcentage_charge_traitement,
        'Nbre_US_clôturés / Nbre_US_total': Avancement1,
        'Temps_consommé/Temps_estimé': Avancement2
    }
    
    # Insérer chaque KPI accompagné de son nom de projet dans un document distinct
    for key, value in projet_doc.items():
        document = {'Nom KPI': key, 'valeur': value, 'Projet': Projet}
        collection4.insert_one(document)
```
